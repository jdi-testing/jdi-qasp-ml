```python
import selenium
from time import sleep
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import os

def maximize_window(driver=None):
    S = lambda X: driver.execute_script('return document.body.parentNode.scroll' + X )
    driver.set_window_size(width=S('Width'), height=S('Height')+1000)
    driver.find_element_by_tag_name('body')
    print('Window maximized')

options = Options()
options.add_argument('--disable-gpu')  # Last I checked this was necessary.
options.add_argument('--skip-js-errors')
options.set_capability("platformName", "windows")
options.add_argument("--start-maximized")
if headless:
    options.add_argument('--headless')

chrome_driver_path = os.path.join(os.getcwd(), 'chromedriver.exe')

if os.path.exists(chrome_driver_path):
    driver = selenium.webdriver.Chrome(executable_path=chrome_driver_path, options=options)
else:
    raise FileNotFoundError(f"chromedriver.exe  {os.getcwd()}")
    
driver.set_window_size(width=1920, height=1080)

driver.get('https://lenta.ru')

sleep(3)
html_e = driver.find_element_by_tag_name('html')
sleep(3)
html_e.send_keys(Keys.CONTROL, Keys.END)
sleep(3)
html_e.send_keys(Keys.CONTROL, Keys.END)
sleep(3)
maximize_window(driver=driver)
sleep(3)
html_e.send_keys(Keys.CONTROL, Keys.END)
sleep(3)
```

```python
a = driver.find_elements_by_tag_name('a')[10]
a
```

```python
GET_ALL_ATTRIBUTES_JS ="""
    var items = {}; 
    for (index = 0; index < arguments[0].attributes.length; ++index) { 
        items[arguments[0].attributes[index].name] = arguments[0].attributes[index].value 
    }; 
    return items;
""".replace('\n','')


def get_all_attributes(driver, web_element):
    try:
        attr_dict = driver.execute_script(GET_ALL_ATTRIBUTES_JS, web_element)
        return attr_dict
    except Exception as ex:
        print(ex)
        return dict()
```

```python
get_all_attributes(driver=driver, web_element=a)
```

```python
[ get_all_attributes(driver=driver, web_element=e)  for e in tqdm(driver.find_elements_by_xpath('//*'))]
```
